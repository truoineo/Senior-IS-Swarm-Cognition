```python
from models import UnicellularModel

model = UnicellularModel()

for _ in range(101):
    model.step()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = model.datacollector.get_model_vars_dataframe()
data_reset = data.reset_index()  # Makes index a column called 'index'

# Plot multiple metrics
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.lineplot(data=data_reset, x='index', y='Population', ax=axes[0, 0])
axes[0, 0].set_title('Population Over Time')
axes[0, 0].set_xlabel('Step')

sns.lineplot(data=data_reset, x='index', y='Mean Energy', ax=axes[0, 1])
axes[0, 1].set_title('Mean Energy Over Time')
axes[0, 1].set_xlabel('Step')

sns.lineplot(data=data_reset, x='index', y='Total Resources Collected', ax=axes[1, 0])
axes[1, 0].set_title('Resources Collected Over Time')
axes[1, 0].set_xlabel('Step')

sns.lineplot(data=data_reset, x='index', y='Survival Rate', ax=axes[1, 1])
axes[1, 1].set_title('Survival Rate Over Time')
axes[1, 1].set_xlabel('Step')

plt.tight_layout()
plt.show()
```
